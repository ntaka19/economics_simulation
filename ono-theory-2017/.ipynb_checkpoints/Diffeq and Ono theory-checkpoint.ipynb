## 1-2 Runge-Kutta 法

shows better fit compared to euler method

```python

tlist=[]
ylist=[]
t=0

dt=0.1

#初期値
t=0
y_old=1
y_new=1

trial=10

#dy/dt=f(t,y)
def f(t,y):
    return y

for time in range(int(trial/dt)):
    tlist.append(t)
    ylist.append(y_new)
    
    k1=f(t,y_old)
    k2=f(t+dt/2,y_old+dt*k1/2)
    k3=f(t+dt/2,y_old+dt*k2/2)
    k4=f(t+dt,y_old+dt*k3)

    y_new=y_old+dt*(k1+2*k2+2*k3+k4)/6
    y_old=y_new
    
    t+=dt
    

#plt.ylim(10, 100)
plt.plot(tlist,ylist,'g--' ,label="Runge-kutta")

x=np.arange(trial+1)
plt.plot(np.exp(x),'r^', label="$e^{x}$")

plt.title("y=$e^{t}$ ",fontsize=20)
plt.xlabel("t",fontsize=20)
plt.ylabel("y",fontsize=20)

plt.legend()
plt.show()
```

## 1-3 Lorentz Attractor



```python
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from matplotlib.font_manager import FontProperties
from pylab import rcParams
rcParams['figure.figsize'] = 10,10

rho=0.04
alpha=0.03
beta=0.000083

tlist=[]
xlist=[]
ylist=[]
zlist=[]

#dm/dt=f
#dc/dt=g

p=10
r=28
b=8/3

def f(t,x,y,z):
    #return -alpha*(c(t)/y-1)*m
    return -p*x+p*y
    #return a*x+b*y

def g(t,x,y,z):
    return -x*z+r*x-y

def h(t,x,y,z):
    return x*y-b*z

#連立微分方程式　Runge-Kutta
def RK(f,g,h,finaltime,tscale):
    
    dt = 0.01
    #計算の精度
    
    x_old=1
    y_old=1
    z_old=1
    
    tt=0
    
    for time in range(int(finaltime/dt)+1):
        #trial/dt は時刻trialまでにdt刻みで何回計算させるかという意味
        if(tt%tscale==0):
            tlist.append(round(tt,1))
            xlist.append(x_old)
            ylist.append(y_old)
            zlist.append(z_old)
            #clist.append(round(c(tt)/1,3))

        k11=f(tt,x_old,y_old,z_old)*dt
        k12=g(tt,x_old,y_old,z_old)*dt
        k13=h(tt,x_old,y_old,z_old)*dt

        k21=f(tt+dt/2,x_old+dt*k11/2,y_old+dt*k12/2,z_old+dt*k13/2)*dt
        k22=g(tt+dt/2,x_old+dt*k11/2,y_old+dt*k12/2,z_old+dt*k13/2)*dt
        k23=h(tt+dt/2,x_old+dt*k11/2,y_old+dt*k12/2,z_old+dt*k13/2)*dt

        k31=f(tt+dt/2,x_old+dt*k21/2,y_old+dt*k22/2,z_old+dt*k23/2)*dt
        k32=g(tt+dt/2,x_old+dt*k21/2,y_old+dt*k22/2,z_old+dt*k23/2)*dt
        k33=h(tt+dt/2,x_old+dt*k21/2,y_old+dt*k22/2,z_old+dt*k23/2)*dt

        k41=f(tt+dt,x_old+dt*k31,y_old+dt*k32,z_old+dt*k33/2)*dt
        k42=g(tt+dt,x_old+dt*k31,y_old+dt*k32,z_old+dt*k33/2)*dt
        k43=h(tt+dt,x_old+dt*k31,y_old+dt*k32,z_old+dt*k33/2)*dt
        
        x_new=x_old+dt*(k11+2*k21+2*k31+k41)/6
        y_new=y_old+dt*(k12+2*k22+2*k32+k42)/6
        z_new=z_old+dt*(k13+2*k23+2*k33+k43)/6
  
        x_old=x_new
        y_old=y_new
        z_old=z_new

        tt=round(tt+dt,1)
    
#return ylist
RK(f,g,h,1000,0.1)

#print("t={tl}".format(tl=tlist))
#print("x={tl}".format(tl=xlist))
#print("y={tl}".format(tl=ylist))
#print("z={tl}".format(tl=zlist))

#plt.xlim(0, 300)
sns.set_style("darkgrid")

fig = sns.mpl.pyplot.figure()
ax = fig.add_subplot(111)
ax.ticklabel_format(useOffset=False)
ax.plot(xlist,ylist,'blue',label="Lorenz Attractor")

plt.title("Lorenz Attractor",fontsize="20")
plt.xlabel("x(t)", fontsize="16")
plt.ylabel("y(t)",fontsize="16")

#for i,txt in enumerate(tlist):
#    ax.annotate("t={tl}".format(tl=txt),(mlist[i],clist[i]))

plt.legend()
plt.show()
```

```python
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.font_manager import FontProperties
from pylab import rcParams
rcParams['figure.figsize'] = 15,10
#figuresizeを大きくする

trial=100
#matplotlib.rcParams.update({'font.size': 10})
#rho=0.039959091
rho=0.04
alpha=0.03
beta=0.000083
#beta=0.00007

y=550
c0=350#兆
m0=50


tlist=[]

def func(alpha,beta,rho,y,c0,m0,mlist,clist,check): 
    #dm/dt=f
    #dc/dt=g
    def f(t,m,c):
        #return -alpha*(c(t)/y-1)*m
        return -alpha*(c/y-1)*m

    def g(t,m,c):
        if(check==0):
            temp=0
            return temp
        else:
            return (beta-alpha/y)*c*c-(rho-alpha)*c

    #連立微分方程式　Runge-Kutta
    def RK(f,g,finaltime,tscale):
        dt = 0.1
        m_old=m0    
        c_old=c0    
        tt=0

        for time in range(int(finaltime/dt)+1):
            #trial/dt は時刻trialまでにdt刻みで何回計算させるかという意味
            if(tt%tscale==0):
                if (len(tlist)<=finaltime/tscale):
                    tlist.append(round(tt,1))
                    
                mlist.append(m_old)
                clist.append(c_old)
                #clist.append(round(c(tt)/1,3))

            k11=f(tt,m_old,c_old)
            k12=g(tt,m_old,c_old)

            k21=f(tt+dt/2,m_old+dt*k11/2,c_old+dt*k12/2)
            k22=g(tt+dt/2,m_old+dt*k11/2,c_old+dt*k12/2)

            k31=f(tt+dt/2,m_old+dt*k21/2,c_old+dt*k22/2)
            k32=g(tt+dt/2,m_old+dt*k21/2,c_old+dt*k22/2)

            k41=f(tt+dt,m_old+dt*k31,c_old+dt*k32)
            k42=g(tt+dt,m_old+dt*k31,c_old+dt*k32)

            m_new=m_old+dt*(k11+2*k21+2*k31+k41)/6
            c_new=c_old+dt*(k12+2*k22+2*k32+k42)/6

            m_old=m_new
            c_old=c_new

            tt=round(tt+dt,1)

    #RK(f,g,finaltime,tscale)
    RK(f,g,200,10)

######FUNCTION########    
    
NUM=5
mlist = [[] for j in range(NUM)]
clist = [[] for j in range(NUM)]
#alpha,beta,rho,y,c0,m0,mlist,clist,check
func(0.03,0.000083,0.04,550,350,50,mlist[0],clist[0],0)
func(0.03,0.000080,0.04,550,350,50,mlist[1],clist[1],1)
func(0.03,0.000085,0.04,550,350,50,mlist[2],clist[2],1)
func(0.03,0.000086,0.04,550,350,50,mlist[3],clist[3],1)
func(0.03,0.00007,0.04,550,350,50,mlist[4],clist[4],1)

#print("t={tl}".format(tl=tlist))
#print("m={tl}".format(tl=mlist))
#print("c={tl}".format(tl=clist))
#plt.xlim(0, 300)
#plt.plot(m(tlist),c(tlist),'r' ,label="c-m")
fig=plt.figure()
ax = fig.add_subplot(111)
ax.ticklabel_format(useOffset=False)

ax.plot(mlist[0],clist[0],'r',label=r"(K) $C_{0}=0$, $\alpha$=0.03, $\beta$=0.000083, $\rho$=0.04, y=550")
ax.plot(mlist[1],clist[1],'b',label=r"(K) $\alpha$=0.03, $\beta$=0.000080, $\rho$=0.04, y=550")
ax.plot(mlist[2],clist[2],'g',label=r"(K) $\alpha$=0.03, $\beta$=0.000085, $\rho$=0.04, y=550")
ax.plot(mlist[3],clist[3],'k',label=r"(K) $\alpha$=0.03, $\beta$=0.000086, $\rho$=0.04, y=550")
ax.plot(mlist[4],clist[4],'y',label=r"(N) $\alpha$=0.03, $\beta$=0.00007, $\rho$=0.04, y=550")
#$\alpha$=$3.0×10^{-2}$

plt.title(r"Ono's c-m field and paths (u(c)=ln(c), v(m)=$\beta$m)",fontsize="20")
plt.xlabel("m(t) Trillion yen", fontsize="16")
plt.ylabel("c(t) Trillion yen",fontsize="16")
plt.text(1100, 700, "t:year", fontsize="16")

plt.text(1100,650,r"$\dotc= (v'(m)-\alpha/y)c^{2}-(\rho+\pi)c$ (=0 when $C_{0}=0$)",fontsize="16")
plt.text(1100,600,r"$\dotm= -\pi m$",fontsize="16")
plt.text(1100,550,r"$\pi= -\alpha(c/y-1)$",fontsize="16")


tstamp=5
#tstamp*timescale=timescale
for i,txt in enumerate(tlist[::tstamp]):
    for t in range(NUM):
        ax.annotate("t={tl}".format(tl=txt),(mlist[t][tstamp*i],clist[t][tstamp*i]))
        #ax.annotate("t={tl}".format(tl=txt),(mlist1[t],clist1[i]))
        #ax.annotate("t={tl}".format(tl=txt),(mlist2[],clist2[i]))
        
plt.legend(prop={'size': 16})
plt.show()


fig.savefig('onos-cm-field.png')   # save the figure to file
```
