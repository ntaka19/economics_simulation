# 第7章　不況の動学経路

## 不況均衡経路(backward)

```python
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.font_manager import FontProperties
from pylab import rcParams
rcParams['figure.figsize'] = 12,10

tlist=[]
eps=0.00001
alpha=0.03
beta=0.00008

beta2=0.0004
rho=0.04
y=550

def cbound(m):
    return y*(rho-alpha)/((beta+beta2/m)*y-alpha)

def func(alpha,beta,rho,y,c0,m0,mlist,clist,check): 
    #dm/dt=f
    #dc/dt=g
    def f(t,m,c):
        return -alpha*(c/y-1)*m

    def g(t,m,c):
        return (beta+beta2/m-alpha/y)*c*c-(rho-alpha)*c

    #連立微分方程式　Runge-Kutta
    def RK(f,g,finaltime,tscale):
        dt = 0.01
        m_old=m0    
        c_old=c0    
        tt=finaltime

        for time in range(int(finaltime/dt)+1):
            #trial/dt は時刻trialまでにdt刻みで何回計算させるかという意味
            if(tt%tscale==0):
                if (len(tlist)<=finaltime/tscale):
                    tlist.append(round(tt,3))
                    
                mlist.append(m_old)
                clist.append(c_old)
                
            k11=f(tt,m_old,c_old)*dt
            k12=g(tt,m_old,c_old)*dt

            k21=f(tt-dt/2,m_old-k11/2,c_old-k12/2)*dt
            k22=g(tt-dt/2,m_old-k11/2,c_old-k12/2)*dt

            k31=f(tt-dt/2,m_old-k21/2,c_old-k22/2)*dt
            k32=g(tt-dt/2,m_old-k21/2,c_old-k22/2)*dt

            k41=f(tt-dt,m_old-k31,c_old-k32)*dt
            k42=g(tt-dt,m_old-k31,c_old-k32)*dt

            km=(k11+2*k21+2*k31+k41)/6
            kc=(k12+2*k22+2*k32+k42)/6
                     
            #収束条件
            if((km < eps) and (kc <eps)):
                break
                
            else:
                m_new=m_old-km
                c_new=c_old-kc
                m_old=m_new
                c_old=c_new
                tt=round(tt-dt,3)
    #RK(f,g,finaltime,tscale)
    RK(f,g,200,1)

######FUNCTION########    
NUM=1
mlist = [[] for j in range(NUM)]
clist = [[] for j in range(NUM)]

##m(T), c(T)の設定
#alpha,beta,rho,y,c0,m0,mlist,clist,check
#func(0.03,0.00008,0.04,550,360,175,mlist[0],clist[0],1)

###最終の値###
c0=370
m0=175
func(alpha,beta,rho,y,c0,m0,mlist[0],clist[0],1)

fig=plt.figure()
ax = fig.add_subplot(111)
ax.ticklabel_format(useOffset=False)
ax.plot(mlist[0],clist[0],'r',label=r"(均衡経路)$\alpha$=0.03, $\beta$=0.00007, $\rho$=0.04, y=550")

x=np.arange(300)
ax.plot(x,list(map(cbound,x)),'b--', label=r"$\dot c=0$")
ax.plot(x,[y for j in range(len(x))],'y--',label=r"$\dot m=0$")

plt.title(r"不況均衡経路($\alpha < \rho < \beta y$)",fontsize="20")
plt.xlabel("m(t) Trillion yen", fontsize="16")
plt.ylabel("c(t) Trillion yen",fontsize="16")

tstamp=100
tlast=tlist[len(tlist)-1]

#tstamp*timescale=timescale
for i,txt in enumerate(tlist[::tstamp]):
    if (i>1 or i==0):
        for t in range(NUM):
            ax.annotate("t={tl}".format(tl=txt-tlast),(mlist[t][tstamp*i],clist[t][tstamp*i]))
#print(mlist[0][1000],clist[0][1000])
#"t={tl}".format(tl=tlist)
print("m(0)={m0}, c(0)={c0}".format(m0=mlist[0][len(mlist[0])-1],c0=clist[0][len(clist[0])-1]))
print("m(T)={m0}, c(T)={c0}".format(m0=mlist[0][0],c0=clist[0][0]))


plt.text(75,250,r"u(c)=ln(c)",fontsize="16")
plt.text(75,220,r"$v(m)=\beta_{1} m+\beta_{2}ln(m) \quad (\beta_{1}=0.00008, \beta_{2}=0.0004$)",fontsize="16")
plt.legend(prop={'size': 16},loc='lower right')
plt.ylim(0,800)
plt.show()

fig.savefig('onos-cm-field(backward)1.png')   # save the figure to file
```

```python
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.font_manager import FontProperties
from pylab import rcParams
rcParams['figure.figsize'] = 12,10

#dtと収束条件、0.001/mの値などでどこで止まるかは変わってしまう。この問題はどう対処すべきか。
tlist=[]
eps=0.00001
#収束条件を弱めると理論的な経路から外れてしまう。
alpha=0.045
beta=0.00007
beta2=0.0004
rho=0.042
y=550

#c\dot=0の境界線
def cbound(m):
    return y*(rho-alpha)/((beta+beta2/m)*y-alpha)

def func(alpha,beta,rho,y,c0,m0,mlist,clist,travlist,finaltime): 
    #dm/dt=f
    #dc/dt=g
    def f(t,m,c):
        return -alpha*(c/y-1)*m

    def g(t,m,c):
        return (beta+beta2/m-alpha/y)*c*c-(rho-alpha)*c

    #連立微分方程式　Runge-Kutta
    def RK(f,g,finaltime,tscale):
        dt = 0.01
        m_old=m0    
        c_old=c0    
        tt=0

        for time in range(int(finaltime/dt)+1):
            #trial/dt は時刻trialまでにdt刻みで何回計算させるかという意味
            if(tt%tscale==0):
                if (len(tlist)<=finaltime/tscale):
                    tlist.append(round(tt,3))
                    
                trav=(1/c_old)*(m_old+10)*np.exp(-rho*tt)
                travlist.append(round(trav,9))    
                mlist.append(m_old)
                clist.append(c_old)

            k11=f(tt,m_old,c_old)*dt
            k12=g(tt,m_old,c_old)*dt
            k21=f(tt+dt/2,m_old+k11/2,c_old+k12/2)*dt
            k22=g(tt+dt/2,m_old+k11/2,c_old+k12/2)*dt
            k31=f(tt+dt/2,m_old+k21/2,c_old+k22/2)*dt
            k32=g(tt+dt/2,m_old+k21/2,c_old+k22/2)*dt
            k41=f(tt+dt,m_old+k31,c_old+k32)*dt
            k42=g(tt+dt,m_old+k31,c_old+k32)*dt

            km=(k11+2*k21+2*k31+k41)/6
            kc=(k12+2*k22+2*k32+k42)/6
                     
            #収束条件
            if((km < eps) and (kc <eps)):
                break
                
            else:
                m_new=m_old+(k11+2*k21+2*k31+k41)/6
                c_new=c_old+(k12+2*k22+2*k32+k42)/6
                m_old=m_new
                c_old=c_new
                tt=round(tt+dt,3)
                #少数第3桁までとする
    
    #RK(f,g,finaltime,tscale)
    RK(f,g,finaltime,1)
    
######FUNCTION########    
NUM=3
mlist = [[] for j in range(NUM)]
clist = [[] for j in range(NUM)]
travlist = [[] for j in range(NUM)]

#alpha,beta,rho,y,c0,m0,mlist,clist,check
func(alpha,beta,rho,y,483.87446066184566,60.81762889767034,mlist[0],clist[0],travlist[0],100)
func(alpha,beta,rho,y,490,62,mlist[1],clist[1],travlist[1],100)
func(alpha,beta,rho,y,483,61,mlist[2],clist[2],travlist[2],40)

#均衡解に到達しうる初期値 m0,c0
fig=plt.figure()
ax = fig.add_subplot(111)
ax.ticklabel_format(useOffset=False)

ax.plot(mlist[0],clist[0],'r',label=r"均衡経路(m(0)=60.82,c(0)=483.87)")
ax.plot(mlist[1],clist[1],'g',label=r"恒常的インフレ経路")
ax.plot(mlist[2],clist[2],'m',label=r"恒常的デフレ経路")

x=np.arange(300)
ax.plot(x,list(map(cbound,x)),'b--',label=r"$\dot c=0$")
ax.plot(x,[y for j in range(len(x))],'y--',label=r"$\dot m=0$")

plt.title(r"不況均衡経路が連続無限個あるケース($\alpha > \rho > \beta y$)",fontsize="20")
plt.xlabel("m(t) Trillion yen", fontsize="16")
plt.ylabel("c(t) Trillion yen",fontsize="16")

#plt.text(1100, 700, "t:year", fontsize="16")
#plt.text(1100,650,r"$\dotc= (v'(m)-\alpha/y)c^{2}-(\rho+\pi)c$ (=0 when $C_{0}=0$)",fontsize="16")
#plt.text(1100,600,r"$\dotm= -\pi m$",fontsize="16")
#plt.text(1100,550,r"$\pi= -\alpha(c/y-1)$",fontsize="16")

######グラフのラベル#####
tstamp=100
tlast=len(tlist)-1
for i,txt in enumerate(tlist[::tstamp]):
    for t in range(1):
        if (i<=2 or i==tlast/tstamp):
            ax.annotate("t={tl}".format(tl=txt),(mlist[t][tstamp*i],clist[t][tstamp*i]),fontsize="14") 
            print("time={nn} trav={tl}".format(nn=txt,tl=travlist[t][tstamp*i]))

tlast=len(clist[1])-1
plt.legend(prop={'size': 16})
plt.ylim(100,1000)
plt.show()
fig.savefig('不況均衡連続無限個.png')   # save the figure to file
```
